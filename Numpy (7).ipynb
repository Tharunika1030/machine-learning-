```python
import numpy as np
a = np.array([1., 2., 3., 4.])
b = np.array([5., 6., 7., 8.])
print(a + b) # [ 6 8 10 12]
print(a - b) # [-4 -4 -4 -4]
print(a * b) # [ 5 12 21 32]
print(a / b) # [0.2 0.33333333 0.42857143 0.5 ]
print(a ** b) # [ 1 64 2187 65536]
```

```python
arr1 = np.array([[1, 2], [3, 4]])
arr2 = np.array([[2, 1], [3, 5]])
greater_than = arr1 > arr2
less_than = arr1 < arr2
equal_to = arr1 == arr2
print("Greater than:\n", greater_than)
print("\nLess than:\n", less_than)
print("\nEqual to:\n", equal_to)
arr3 = np.array([1, 2, 3])
arr4 = np.array([1, 2, 3])
arr5 = np.array([1, 2, 4])
result1 = np.array_equal(arr3, arr4) # True
result2 = np.array_equal(arr3, arr5) # False
print("\nArray-wise comparison 1:", result1)
print("Array-wise comparison 2:", result2)
```

```python
x = np.arange(10)
print("x= ",x)

print("\n x[:5]= ",x[:5]) 
print("\n x[5:]= ",x[5:]) 
print("\n x[4:7]= ",x[4:7]) 
print("\n x[::2]= ",x[::2]) 
print("\n x[1::2]= ",x[1::2]) 
print("\n x[::-1]= ",x[::-1]) 
print("\n x[5::-2]= ",x[5::-2]) 
```

```python
x = np.arange(10)
print(x)
x[5:] = 2
print("\n After broadcasting1",x)
x[::2] = 100
print("\n After broadcasting2",x)
```
