```python
from sklearn.datasets import load_wine
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
import numpy as np
X, y = load_wine(return_X_y=True)
X_scaled = StandardScaler().fit_transform(X)
pca_2d = PCA(n_components=2)
X_pca = pca_2d.fit_transform(X_scaled)
plt.figure(figsize=(8,6))
plt.scatter(X_pca[:,0], X_pca[:,1], c=y, cmap='rainbow', edgecolor='k', s=50)
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.title('PCA - Wine Dataset (2 Components)')
plt.show()
print("Explained Variance Ratio for 2 PCs:", pca_2d.explained_variance_ratio_)
print("Total variance explained (~):", round(sum(pca_2d.explained_variance_ratio_),2))
pca_full = PCA()
cum_variance = np.cumsum(pca_full.fit(X_scaled).explained_variance_ratio_)
plt.figure(figsize=(8,6))
plt.plot(range(1, len(cum_variance)+1), cum_variance, marker='o')
plt.axhline(0.95, color='r', linestyle='--', label='95% Explained Variance')
plt.xlabel('Number of Principal Components')
plt.ylabel('Cumulative Explained Variance')
plt.title('Cumulative Variance Explained by PCs')
plt.legend()
plt.show()
num_components_95 = np.argmax(cum_variance >= 0.95) + 1
print("Number of components needed for 95% variance:", num_components_95)
```
